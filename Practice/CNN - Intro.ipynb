```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImg(img_src,title="Image"):
    
    #img_src = cv2.resize(img,(100,100))
    if(title=='GRAY'):
        plt.imshow(img_src,cmap='gray')
    else:
        plt.imshow(img_src)
    plt.axis('off')
    #plt.style.use('seaborn')
    plt.title(title+str(img_src.shape))
    plt.show()
```

```python
img = cv2.imread('./1.jpeg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
drawImg(img,"RGB")
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
drawImg(img_gray,"GRAY")
```

```python
#Hardcoded filters
def convolution(img,img_filter):
    W = img.shape[0]
    H = img.shape[1]
    
    F = img_filter.shape[0] #3
    new_img = np.zeros((W-F+1,W-F+1))
    
    for row in range(W-F+1):
        for col in range(H-F+1):
            for i in range(F):
                for j in range(F):
                    new_img[row][col] += img[row+i][col+j]*img_filter[i][j]
                
            if(new_img[row][col]>255):
                new_img[row][col] = 255
            elif(new_img[row][col]<0):
                new_img[row][col] = 0
            
    
    return new_img #Activation Map
```

```python
blur_filter = np.ones((3,3))/9.0

output1 = convolution(img_gray,blur_filter)
```

```python
drawImg(output1,"GRAY")
drawImg(img_gray,"GRAY")
```

```python
edge_filter = np.array(([1,0,-1],
                       [1,0,-1],
                       [1,0,-1]))
```

```python
output2 = convolution(img_gray,edge_filter)
```

```python
drawImg(output2,"GRAY")
```
