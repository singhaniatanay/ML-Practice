```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('2_class_data.csv')
```

```python
print(df)
```

```python
X = np.array(df[['x1','x2']])
Y = np.array(df['y'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
LR_classifier = LogisticRegression()
DT_classifier = DecisionTreeClassifier()
SVM_classifier = SVC(kernel='rbf',gamma=200)
NN_classifier = MLPClassifier()
```

```python
LR_classifier.fit(X,Y)
DT_classifier.fit(X,Y)
SVM_classifier.fit(X,Y)
NN_classifier.fit(X,Y)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.title('Original')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=LR_classifier.predict(X))
plt.title('Logistic Regression')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=DT_classifier.predict(X))
plt.title('Decision Tree')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=SVM_classifier.predict(X))
plt.title('SVM')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=NN_classifier.predict(X))
plt.title('Nueral Networks')
plt.show()
```
