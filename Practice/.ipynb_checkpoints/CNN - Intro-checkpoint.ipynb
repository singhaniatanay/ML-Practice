```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImg(img_src,title="Image"):
    img = cv2.imread(img_src)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#     img = cv2.resize(img,(100,100))
    plt.imshow(img)
    plt.axis('off')
    plt.style.use('seaborn')
    plt.title(title+str(img.shape))
    plt.show()
    
```

```python
drawImg('./1.jpeg')
```
